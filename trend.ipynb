```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
%matplotlib inline
trends = TrendReq()
```

Here I will be analyzing the Google search trends on the queries based on "Shah rukh khan”, so let’s create a DataFrame of the top 10 countries which search for "SRK" on Google

```python
trends.build_payload(kw_list=['SRK'])  #it will create a list of key word that u want to seach in google trends
data = trends.interest_by_region() #it will search this keyword by particular region
#it scale the search keyword out of 100 where 100 signifies the most number of search
data = data.sort_values(by= 'SRK', ascending=False)
data = data.head(10)
print(data)
```

```python
data.plot.bar(color= 'steelblue', figsize=(15,10) )
plt.xlabel("Region")
plt.ylabel('Search Score')
plt.title("Search Score for SRK")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data.plot.area(alpha= 0.35, figsize=(15,10),stacked= False)
```

Now lets look at this search on a timely basis how frequent it increases or descreases on google

```python
trends = TrendReq(hl= 'en-US', tz=360)
trends.build_payload(kw_list=['SRk'])
data2 = trends.interest_over_time()

data2 = data2.head(100)
```

```python
data2.plot(kind= 'line', figsize= (15,10), color= 'r', marker= 'o')
plt.title('Total Google Searches for SRK', fontWeight= 'bold')
plt.xlabel('Year')
plt.ylabel('Total count')
plt.show()
```

So we can see a huge drop in the seaches about SRK on google in 2019 july and so on.
